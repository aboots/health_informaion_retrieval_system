## tfidf - fasttext - transformer models 

```python
class QueryHandler:
    def __init__(self):
        self.transformer_model = TransformerEmb()
        self.tfidf_model = TfidfModel()
        self.fasttext_model = FastTextEmb()
    
    def print_similars(self, query, k=10):
        print('tfidf model:')
        self.tfidf_model.print_similars(query, k)
        print('transformer model:')
        self.transformer_model.print_similars(query, k)
        print('fasttext model:')
        self.fasttext_model.print_similars(query, k)
```

```python
query_handler = QueryHandler()
```

```python
query_handler.print_similars('ریزش مو')
```

```python
query_handler.print_similars('ویروس کرونا')
```

```python
query_handler.print_similars('کمبود خواب')
```

```python
query_handler.print_similars('سکته قلبی و مغزی')
```

```python
query_handler.print_similars('اعتیاد به مواد مخدر')
```

```python
query_handler.print_similars('نوشیدن آب')
```

```python
query_handler.print_similars('میوه و سبزیجات')
```

```python
query_handler.print_similars('کامپوزیت دندان')
```

```python
query_handler.print_similars('افسردگی و اضطراب')
```

```python
query_handler.print_similars('کبد چرب')
```
